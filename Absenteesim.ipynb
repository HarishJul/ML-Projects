```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
emp = pd.read_csv("Absenteeism_at_work.csv", sep = ';')
print(emp.shape)
```

```python
col_names = ['id','reason_for_absence','month_of_absence','day_of_the_week','seasons','transportation_expense','distance_from_residence_to_work',
            'service_time','age','work_load_average/day','hit_target','disciplinary_failure','education','son','social_drinker','social_smoker',
             'pet','weight','height','body_mass_index','absenteeism_time_in_hours']

emp.columns = col_names
emp.head()
```

```python
emp.info()
```

```python
cat_cols = ['id','reason_for_absence','month_of_absence','day_of_the_week','seasons']

for col in cat_cols:
    emp[col] = emp[col].astype('object')    
```

```python
emp.isna().sum()
```

```python
emp.describe()
```

```python
emp.head()
```

```python
emp['id'].unique()
```

```python
_ = emp.groupby('id')['absenteeism_time_in_hours'].mean().plot.bar()
fig = plt.gcf()
fig.set_size_inches(12,6)
plt.show()
```

```python
#For what reason there are more absentees
_ = emp.groupby('reason_for_absence')['id'].count().plot.bar()
fig = plt.gcf()
fig.set_size_inches(12,6)
plt.show()
```

```python
_ = emp.groupby('reason_for_absence')['absenteeism_time_in_hours'].sum().plot.bar()
fig = plt.gcf()
fig.set_size_inches(12,6)
plt.show()
```

```python
_ = emp.groupby('month_of_absence')['id'].count().plot.bar()
fig = plt.gcf()
fig.set_size_inches(12,6)
plt.show()
```

```python
_ = emp.groupby('month_of_absence')['absenteeism_time_in_hours'].sum().plot.bar()
fig = plt.gcf()
fig.set_size_inches(12,6)
plt.show()
```

```python
_ = emp.groupby('day_of_the_week')['absenteeism_time_in_hours'].sum().plot.bar()
fig = plt.gcf()
fig.set_size_inches(12,6)
plt.show()
```

```python
_ = emp.groupby('seasons')['absenteeism_time_in_hours'].sum().plot.bar()
fig = plt.gcf()
fig.set_size_inches(10,5)
plt.show()
```

```python
_ = emp[emp['son'] == 0]['absenteeism_time_in_hours'].plot.hist(edgecolor='black', linewidth=1.2, bins = 20)
fig = plt.gcf()
fig.set_size_inches(10,5)
plt.show()
```

```python
_ = emp[emp['son'] != 0]['absenteeism_time_in_hours'].plot.hist(edgecolor='black', linewidth=1.2, bins = 20)
fig = plt.gcf()
fig.set_size_inches(10,5)
plt.show()
```

```python
num_col = ['transportation_expense','distance_from_residence_to_work','service_time','age','work_load_average/day','hit_target',
          'weight','height','body_mass_index', 'absenteeism_time_in_hours']

_ = emp[num_col].hist(edgecolor='black', linewidth=1.2)
fig = plt.gcf()
fig.set_size_inches(18,9)
plt.show()
```

```python
corr = emp[num_col].corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    ax = sns.heatmap(corr, mask=mask, vmax=.3, square=True)
fig = plt.gcf()
fig.set_size_inches(14,7)
plt.show()
```
